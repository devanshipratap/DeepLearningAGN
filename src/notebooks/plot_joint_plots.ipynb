```python
"""
Joint plots useful for visualizing histograms inline with scatter plots
"""
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading data file 
DF = pd.read_csv('../../data/merged_simulated.csv')
DF = DF[(DF.Mass_ground_truth != 0)]
```

```python
# Joint plot for mass ground truth vs mass prediction 
sns.jointplot(DF['Mass_ground_truth'], DF['Mass_prediction'], color='darkblue', kind='reg',
              scatter_kws={'s': 10, 'alpha': .5}, line_kws={'color': 'black'})
plt.xlabel('Mass Ground Truth')
plt.ylabel('Mass Prediction')
plt.annotate('Pearson = 0.193\nSpearman = 0.200\nP-Value = 9.19e-17', xy=(.1, .9),
             xycoords='axes fraction', size=10)
plt.show()
```

```python
# Joint plot for z ground truth vs z prediction 
sns.jointplot(DF['z_ground_truth'], DF['z_prediction'], color='darkred', kind='reg',
              scatter_kws={'s': 10, 'alpha': .5}, line_kws={'color': 'black'})
plt.xlabel('Redshift Ground Truth')
plt.ylabel('Redshift Prediction')
plt.annotate('Pearson = 0.834\nSpearman = 0.776\nP-Value = 0', xy=(.1, .9),
             xycoords='axes fraction', size=10)
plt.show()
```
