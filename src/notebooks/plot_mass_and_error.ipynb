```python
"""
Plot mass vs. mass errors
"""
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read .csv file for data to plot
MASS_DF = pd.read_csv('../../data/clean_full_data.csv')
MASS_DF = MASS_DF[(MASS_DF.M != 0)]
```

```python
# Plot mass vs. Mass error
plt.figure(figsize=(8, 5))
sns.set(font='Times New Roman')
sns.set_context('paper')
sns.regplot(x=MASS_DF['M'], y=MASS_DF['ERR'], scatter=True, marker='o', fit_reg=False,
            color='red', scatter_kws={'s': 5})
plt.xlabel('Mass',  fontsize=12)
plt.ylabel('Mass Error',  fontsize=12)
plt.title('M vs Mass error from DR7 catalog',  fontsize=14)
plt.show()
```

```python
# Plot mass error vs count
plt.figure(figsize=(8, 5))
sns.set(font='Times New Roman')
sns.set_context('paper')
sns.distplot(MASS_DF['ERR'], kde=False, color='red')
plt.title('Mass Statistical Analysis',  fontsize=15)
plt.xlabel('Mass Error',  fontsize=12)
plt.ylabel('Count',  fontsize=12)
plt.show()
```
