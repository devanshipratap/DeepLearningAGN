```python
"""
Plot neural network results for redshift of the simulated data set
"""
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_z_simulated_result(file_path):
    """
    Plot neural network results for redshift of the simulated data set
    """

    # Defining the data frame 
    df = pd.read_csv(file_path)

    # Plotting
    plt.figure(figsize=(8, 5))
    sns.set_context('paper')
    sns.set(font='Times New Roman')
    ground_truth = plt.plot(df['z_ground_truth'], df['z_ground_truth'], color='blue', zorder=2)
    scatter = plt.scatter(df['z_ground_truth'], df['z_prediction'], 
                          color='black', alpha=.5, s=10, zorder=1)
    error = plt.errorbar(df['z_ground_truth'], df['z_prediction'], 
                         yerr=df['z_error_prediction'], ls='', ecolor='grey', alpha=.2, zorder=0)
    plt.xlabel('Redshift', fontsize=12)
    plt.ylabel('Redshift', fontsize=12)
    plt.title('Neural Network Results for Redshift', fontsize=14)
    plt.legend((ground_truth, scatter, error), labels=('Ground truth', 'NN prediction', 'Errors'))
    plt.show()
```

```python
plot_z_simulated_result('../../data/z_prediction_simulated.csv')
```
