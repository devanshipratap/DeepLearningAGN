```python
import pandas as pd
from astropy.io import fits
import numpy as np
```

```python
FILE_PATH = '/Users/SnehPandya/Desktop/DeepLearningAGN/data/dr14q_spec_prop_ext.fits'
dr14 = fits.open(FILE_PATH)
```

```python
DATA = dr14[1].data
DATA
```

```python
SDSS_ID = DATA.field('SDSS_ID')
SDSS_NAME = DATA.field('SDSS_NAME')
MJD = DATA.field('MJD')
RA = DATA.field('RA')
DEC = DATA.field('DEC')
MASS = DATA.field('LOG_MBH')
ERR = DATA.field('LOG_MBH_ERR')
Z = DATA.field('REDSHIFT')
MI = DATA.field('MI')
```

```python
X_TRAIN = pd.DataFrame(SDSS_NAME, columns=['SDSS_Name'])
X_TRAIN['ID'] = SDSS_ID
X_TRAIN['MJD'] = MJD
X_TRAIN['ra'] = RA
X_TRAIN['dec'] = DEC
X_TRAIN['Mass'] = MASS
X_TRAIN['z'] = Z
X_TRAIN['ERR'] = ERR
X_TRAIN['M_i'] = MI
```

```python
X_TRAIN.to_csv('/Users/SnehPandya/Desktop/DeepLearningAGN/data/dr14_BH.csv')
```
