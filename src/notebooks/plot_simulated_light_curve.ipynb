```python
"""
Plot a simulated version of an arbitrary Light Curve from Stripe 82 data
Information on data columns found at:
http://faculty.washington.edu/ivezic/macleod/qso_dr7/Southern_format_LC.html
"""
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_light_curve(file_path):
    """
    Plot a simulated light curve
    """

    # Opening a light curve file
    with open(file_path, 'r') as f:
        next(f)
        df = pd.DataFrame(l.rstrip().split() for l in f)

    # Converting string to numeric
    for i in range(15):
        df[i] = pd.to_numeric(df[i])

    # Cleaning unphysical magnitudes
    df_u = df[df[1] > 1]
    df_g = df[df[4] > 1]
    df_r = df[df[7] > 1]
    df_i = df[df[10] > 1]
    df_z = df[df[13] > 1]

    # Plotting
    fig = plt.figure(figsize=(8, 5))
    sns.set(font='Times New Roman')
    sns.set_context('paper')
    u = plt.plot(df_u[0], np.random.normal(df_u[1],df_u[2]), label='u', color='purple', marker='*', ls='')
    g = plt.plot(df_g[3], np.random.normal(df_g[4],df_g[5]), label='g', color='green', marker='*', ls='')
    r = plt.plot(df_r[6], np.random.normal(df_r[7],df_r[8]), label='r', color='red', marker='*', ls='')
    i = plt.plot(df_i[9], np.random.normal(df_i[10],df_i[11]), label='i', color='yellow', marker='*', ls='')
    z = plt.plot(df_z[12], np.random.normal(df_z[13],df_z[14]), label='z', color='blue', marker='*', ls='')
    plt.xlabel('Modified Julian Day [MJD]', fontsize=12)
    plt.ylabel('Magnitude',  fontsize=12)
    plt.title('Simulated Light Curve', fontsize=15)
    plt.gca().invert_yaxis()
    plt.legend((u, g, r, i, z), labels=('u', 'g', 'r', 'i', 'z'))
    plt.show()
```

```python
# Plot a simulated light curve
plot_light_curve('../scripts/light_curves/stripe82/70')
```
